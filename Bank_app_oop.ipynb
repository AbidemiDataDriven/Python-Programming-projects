## Python Object Oriented Programming (OOP)- BANK APPLICATION

```python
## Parant class
class customer():
    def __init__(self, name, age,gender, contact_number, home_address):
        self.name = name
        self.age =age
        self.gender= gender
        self.contact_number =contact_number
        self.home_address = home_address
    def show_customer_details(self):
        print("customer_details are as follows...............")
        print(" ")
        print("Name = ", self.name)
        print("Age =", self.age)
        print("Gender =", self.gender)
        print("Contact Number=", self.contact_number)
        print("Home Address =", self.home_address)

##child class
class Bank(customer):
    def __init__(self, name, age, gender, contact_number, home_address):
        super().__init__(name, age, gender, contact_number, home_address)
        self.balance=0

    def deposit_transaction(self, amount):
        self.amount=amount
        self.balance =self.balance + self.amount
        print("Your account Balance is credited by #", self.amount)
        print("Your accounnt balance is #:", self.balance)

    def withdraw_transaction(self, amount):
        self.amount = amount
        if self.amount > self.balance:
            print("Sorry! Insufficient balance! your available balance #:", self.balance)
        else:
            self.balance = self.balance - self.amount
            print("Your account Balance is debited by #", self.amount)
            print("Your accounnt balance is #:", self.balance)
    def check_balance(self):
        print("Availabe Balance account is #:", self.balance)
```
